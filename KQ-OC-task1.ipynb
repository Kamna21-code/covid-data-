# Daily Cases

```python
# dropping unnecessary columns
cases = data.drop(['Province/State','Country/Region','Lat','Long'],axis=1).sum()
cases =  cases.reset_index(drop= True)
```

```python
# printing daily cases world wide
cases
```

# Increase in Cases

```python
# Defininig a function to calculate increment in cases
def increase(s):
    inc = cases.diff().tolist()
    return inc
    
```

```python
# printing daily increment of cases
increase(cases)
```

# Daily Average of cases

```python
# Defining a function to take average of cases
def avg(cases,k):
    for i in range(0,432,k):
        s=0
        for j in range(i,i+k):
            s=s+cases[j]
        print(int(s/k))
```

```python
# printing average cases
plt_avg =avg(cases,8)
plt_avg
```

# Plotting the data

```python
# plotting cases data with the hhelp of Histogram
num_bins = 20
n, bins, patches = plt.hist(cases, num_bins, facecolor='blue', alpha=0.5)
plt.xlabel('Cases')
plt.ylabel('Daily')
plt.title(r'Daily Cases')

plt.subplots_adjust(left=0.15)
plt.show()
```
